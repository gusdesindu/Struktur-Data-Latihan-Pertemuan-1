1.5 Nama Mahasiswa

```python
mahasiswa = ["Andi", "Budi", "Gata", "Rehan", "Galang"]
print("List Mahasiswa: ")
print(mahasiswa)
```

Tambahan 1 Nama Baru dengan append

```python
mahasiswa.append("Gusde")
print("Setelah ditambahkan nama baru")
print(mahasiswa)
```

Hapus Satu Elemen Menggunakan Remove

```python
mahasiswa.remove("Andi")
print("Setelah menghapus nama Anomali")
print(mahasiswa)
```
