Input Data Awal

```python
arr = []

jumlah = int(input("Masukkan jumlah elemen array: "))

for i in range(jumlah):
    angka = int(input(f"Masukkan angka ke-{i+1}: "))
    arr.append(angka)

print("Array awal:", arr)
```

Fungsi INSERT TENGAH (Manual)

```python
def insert_tengah(arr, posisi, nilai):
    # menambah 1 slot kosong
    arr.append(0)

    # geser elemen ke kanan
    for i in range(len(arr) - 1, posisi, -1):
        arr[i] = arr[i - 1]

    # masukkan nilai
    arr[posisi] = nilai
```

Fungsi DELETE TENGAH (Manual)

```python
def delete_tengah(arr, posisi):
    # geser elemen ke kiri
    for i in range(posisi, len(arr) - 1):
        arr[i] = arr[i + 1]

    # hapus elemen terakhir
    arr.pop()
```

Pemanggilan Fungsi

```python
# Insert data
posisi_insert = int(input("Masukkan posisi insert: "))
nilai_insert = int(input("Masukkan nilai yang akan diinsert: "))

insert_tengah(arr, posisi_insert, nilai_insert)
print("Array setelah insert:", arr)

# Delete data
posisi_delete = int(input("Masukkan posisi delete: "))
delete_tengah(arr, posisi_delete)

print("Array setelah delete:", arr)
```
